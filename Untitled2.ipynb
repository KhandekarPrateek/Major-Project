```python
import os
import zipfile
import gdown

import gdown

zip1_url = "https://drive.google.com/uc?id=1P_O4dkE4-IOiSUw_mihAdGJ-F1rGN9PF"
zip2_url = "https://drive.google.com/uc?id=1SYlObf8XLj3S5oumdHWll2F90jls6_p_"

zip1_path = "/content/dataset1.zip"
zip2_path = "/content/dataset2.zip"

# Use --fuzzy to handle redirects
gdown.download(zip1_url, zip1_path, quiet=False, fuzzy=True)
gdown.download(zip2_url, zip2_path, quiet=False, fuzzy=True)


# Extract ZIP files
def extract_zip(zip_path, extract_to):
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_to)

extract_zip(zip1_path, "/content/dataset1")
extract_zip(zip2_path, "/content/dataset2")

print("Extraction complete!")
```

```python
import os
import shutil

# Paths to datasets
augmented_path = "/content/dataset1/Augmented Images"
downloaded_path = "/content/dataset2/Downloaded_Images"
merged_path = "/content/merged_dataset"

# Create merged directory if not exists
os.makedirs(merged_path, exist_ok=True)

# Function to normalize plant names
def normalize_name(name):
    return name.replace("_", " ").replace(".", "").strip().lower()

# Get plant names from both datasets
augmented_plants = {normalize_name(name): name for name in os.listdir(augmented_path)}
downloaded_plants = {normalize_name(name): name for name in os.listdir(downloaded_path)}

# Merge images into a single plant folder
for norm_name in set(augmented_plants.keys()).union(downloaded_plants.keys()):
    # Get actual plant folder names
    aug_folder = augmented_plants.get(norm_name)
    down_folder = downloaded_plants.get(norm_name)

    # Define final merged folder
    merged_folder = os.path.join(merged_path, norm_name)
    os.makedirs(merged_folder, exist_ok=True)

    # Copy images from augmented dataset
    if aug_folder:
        aug_folder_path = os.path.join(augmented_path, aug_folder)
        for img in os.listdir(aug_folder_path):
            shutil.copy(os.path.join(aug_folder_path, img), merged_folder)

    # Copy images from downloaded dataset
    if down_folder:
        down_folder_path = os.path.join(downloaded_path, down_folder)
        for img in os.listdir(down_folder_path):
            shutil.copy(os.path.join(down_folder_path, img), merged_folder)

    print(f"Merged: {norm_name}")

print("✅ Merging complete! Check /content/merged_dataset")
```
