```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://bsi.gov.in/page/en/medicinal-plant-database"

# Send request
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# Check the content
print(soup.prettify()[:2000])  # Prints first 2000 characters of the HTML

```

```python
tables = soup.find_all("table")  # Get all tables
print(len(tables))  # Number of tables found

# Select the correct table (usually first one)
table = tables[0]
```

```python
# Extract table headers
headers = [th.text.strip() for th in table.find_all("th")]
print(headers)  # Check column names

# Extract table rows
rows = []
for tr in table.find_all("tr")[1:]:  # Skip header row
    cells = tr.find_all("td")
    row = [cell.text.strip() for cell in cells]
    rows.append(row)

# Create DataFrame
bsi_df = pd.DataFrame(rows, columns=headers)

# Save to CSV
bsi_df.to_csv("bsi_data.csv", index=False)

print("BSI Data Saved Successfully!")
```

```python
import cv2
import numpy as np
from bing_image_downloader import downloader
import os
import random
from pathlib import Path
import glob

def extract_plant_name(image_path):
    filename = os.path.splitext(os.path.basename(image_path))[0]
    plant_name = filename.replace('_', ' ').replace('-', ' ')
    plant_name = ' '.join(word.capitalize() for word in plant_name.split())
    return plant_name

def enhance_green(image):
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    mask = cv2.inRange(hsv, (35, 25, 25), (85, 255, 255))
    hsv[:, :, 1] = np.where(mask == 255, np.minimum(hsv[:, :, 1] * 1.5, 255), hsv[:, :, 1])
    return cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)

def increase_brightness(image, factor):
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    hsv[:, :, 2] = np.clip(hsv[:, :, 2] * factor, 0, 255)
    return cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)

def rotate_image(image, angle):
    height, width = image.shape[:2]
    center = (width // 2, height // 2)
    rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)
    return cv2.warpAffine(image, rotation_matrix, (width, height))

def flip_image(image, flip_code):
    return cv2.flip(image, flip_code)

def apply_gaussian_blur(image, kernel_size=(5,5)):
    return cv2.GaussianBlur(image, kernel_size, 0)

def add_noise(image):
    noise = np.random.normal(0, 25, image.shape).astype(np.uint8)
    noisy_image = cv2.add(image, noise)
    return noisy_image

def adjust_contrast(image, factor):
    mean = np.mean(image)
    adjusted = (image - mean) * factor + mean
    return np.clip(adjusted, 0, 255).astype(np.uint8)

def apply_color_jitter(image):
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    hsv = hsv.astype(np.float32)
    hsv[:,:,0] = hsv[:,:,0] * random.uniform(0.9, 1.1)
    hsv[:,:,1] = hsv[:,:,1] * random.uniform(0.8, 1.2)
    hsv[:,:,2] = hsv[:,:,2] * random.uniform(0.8, 1.2)
    hsv = np.clip(hsv, 0, 255).astype(np.uint8)
    return cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)

import shutil

def download_reference_images(plant_name, output_dir, num_images=5):
    temp_dir = os.path.join(output_dir, plant_name)  # Bing downloader creates a subfolder

    try:
        downloader.download(
            plant_name,
            limit=num_images,
            output_dir=output_dir,
            adult_filter_off=True,
            force_replace=False,
            timeout=60
        )
        print(f"Successfully downloaded {num_images} images of {plant_name}")

        # Move images from temp_dir to output_dir
        if os.path.exists(temp_dir):
            for file in os.listdir(temp_dir):
                shutil.move(os.path.join(temp_dir, file), output_dir)
            os.rmdir(temp_dir)  # Remove empty subfolder

    except Exception as e:
        print(f"Error downloading images: {str(e)}")



def augment_image(image_path, output_dir):
    Path(output_dir).mkdir(parents=True, exist_ok=True)
    plant_name = extract_plant_name(image_path)
    print(f"Extracted plant name: {plant_name}")
    img = cv2.imread(image_path)
    if img is None:
        raise ValueError(f"Could not read image at {image_path}")
    augmentations = {
        "original": img,
        "green_enhanced": enhance_green(img.copy()),
        "brightness_increased": increase_brightness(img.copy(), 1.5),
        "rotated_90": rotate_image(img.copy(), 90),
        "rotated_180": rotate_image(img.copy(), 180),
        "rotated_270": rotate_image(img.copy(), 270),
        "flipped_horizontal": flip_image(img.copy(), 1),
        "flipped_vertical": flip_image(img.copy(), 0),
        "gaussian_blur": apply_gaussian_blur(img.copy()),
        "noise_added": add_noise(img.copy()),
        "contrast_increased": adjust_contrast(img.copy(), 1.5),
        "color_jittered": apply_color_jitter(img.copy())
    }
    for name, augmented_img in augmentations.items():
        output_path = os.path.join(output_dir, f"{plant_name}_{name}.jpg")
        cv2.imwrite(output_path, augmented_img)
        print(f"Saved {name} augmentation to {output_path}")
    download_reference_images(plant_name, output_dir)

def process_all_images(input_dir, output_dir):
    image_extensions = ['.jpg', '.jpeg', '.png', '.bmp']
    image_files = []
    for ext in image_extensions:
        image_files.extend(glob.glob(os.path.join(input_dir, f'*{ext}')))
    if not image_files:
        print(f"No images found in {input_dir}")
        return
    for image_path in image_files:
        print(f"\nProcessing image: {image_path}")
        try:
            augment_image(image_path, output_dir)
            print(f"Successfully processed {image_path}")
        except Exception as e:
            print(f"Error processing {image_path}: {str(e)}")

def main():
    input_dir = "/content/images"
    output_dir = "/content/combined_images80"
    try:
        process_all_images(input_dir, output_dir)
        print("\nAll images processed successfully!")
    except Exception as e:
        print(f"Error during processing: {str(e)}")
if __name__ == "__main__":
    main()
```
