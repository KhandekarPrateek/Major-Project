```python
import cv2
import numpy as np
import os
import random
from pathlib import Path
import glob
from tqdm import tqdm

def extract_plant_name(image_path):
    filename = os.path.splitext(os.path.basename(image_path))[0]
    plant_name = filename.replace('_', ' ').replace('-', ' ')
    plant_name = ' '.join(word.capitalize() for word in plant_name.split())
    return plant_name

def enhance_green(image):
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    mask = cv2.inRange(hsv, (35, 25, 25), (85, 255, 255))
    hsv[:, :, 1] = np.where(mask == 255, np.minimum(hsv[:, :, 1] * 1.5, 255), hsv[:, :, 1])
    return cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)

def increase_brightness(image, factor):
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    hsv[:, :, 2] = np.clip(hsv[:, :, 2] * factor, 0, 255)
    return cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)

def rotate_image(image, angle):
    height, width = image.shape[:2]
    center = (width // 2, height // 2)
    rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)
    return cv2.warpAffine(image, rotation_matrix, (width, height))

def flip_image(image, flip_code):
    return cv2.flip(image, flip_code)

def apply_gaussian_blur(image, kernel_size=(5,5)):
    return cv2.GaussianBlur(image, kernel_size, 0)

def add_noise(image):
    noise = np.random.normal(0, 25, image.shape).astype(np.uint8)
    noisy_image = cv2.add(image, noise)
    return noisy_image

def adjust_contrast(image, factor):
    mean = np.mean(image)
    adjusted = (image - mean) * factor + mean
    return np.clip(adjusted, 0, 255).astype(np.uint8)

def apply_color_jitter(image):
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV).astype(np.float32)
    hsv[:,:,0] *= random.uniform(0.9, 1.1)
    hsv[:,:,1] *= random.uniform(0.8, 1.2)
    hsv[:,:,2] *= random.uniform(0.8, 1.2)
    hsv = np.clip(hsv, 0, 255).astype(np.uint8)
    return cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)

def augment_image(image_path, output_dir):
    Path(output_dir).mkdir(parents=True, exist_ok=True)
    plant_name = extract_plant_name(image_path)
    img = cv2.imread(image_path)
    if img is None:
        raise ValueError(f"Could not read image at {image_path}")
    plant_output_dir = os.path.join(output_dir, plant_name.replace(' ', '_'))
    Path(plant_output_dir).mkdir(parents=True, exist_ok=True)
    augmentations = {
        "original": img,
        "green_enhanced": enhance_green(img.copy()),
        "brightness_increased": increase_brightness(img.copy(), 1.5),
        "rotated_90": rotate_image(img.copy(), 90),
        "rotated_180": rotate_image(img.copy(), 180),
        "rotated_270": rotate_image(img.copy(), 270),
        "flipped_horizontal": flip_image(img.copy(), 1),
        "flipped_vertical": flip_image(img.copy(), 0),
        "gaussian_blur": apply_gaussian_blur(img.copy()),
        "noise_added": add_noise(img.copy()),
        "contrast_increased": adjust_contrast(img.copy(), 1.5),
        "color_jittered": apply_color_jitter(img.copy())
    }
    for name, augmented_img in tqdm(augmentations.items(), desc=f"Processing {plant_name}"):
        output_path = os.path.join(plant_output_dir, f"{name}.jpg")
        cv2.imwrite(output_path, augmented_img)

def process_all_images(input_dir, output_dir):
    image_extensions = ['.jpg', '.jpeg', '.png', '.bmp']
    image_files = [file for ext in image_extensions for file in glob.glob(os.path.join(input_dir, f'*{ext}'))]
    if not image_files:
        print(f"No images found in {input_dir}")
        return
    for image_path in tqdm(image_files, desc="Processing Images", unit="image"):
        try:
            augment_image(image_path, output_dir)
        except Exception as e:
            print(f"Error processing {image_path}: {str(e)}")

def main():
    input_dir = "images"  # Directory containing your plant images
    output_dir = "augmented_images12"  # Where to save augmented images
    try:
        process_all_images(input_dir, output_dir)
        print("\nAll images processed successfully!")
    except Exception as e:
        print(f"Error during processing: {str(e)}")

if __name__ == "__main__":
    main()
```

```python
import shutil
import os
import time
from tqdm import tqdm

def zip_folder(folder_name, output_name):
    if os.path.exists(folder_name):
        print(f"Zipping {folder_name}...")
        for _ in tqdm(range(100), desc=f"Processing {folder_name}"):
            time.sleep(0.02)  # Simulating progress
        shutil.make_archive(output_name, 'zip', folder_name)
        print(f"{folder_name} has been zipped as {output_name}.zip")
    else:
        print(f"Folder {folder_name} does not exist.")

def main():
    zip_folder("augmented_images12", "augmented_images_12")
    print("Folders zipped successfully!")

if __name__ == "__main__":
    main()
```
